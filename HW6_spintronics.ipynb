```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.constants import physical_constants

m = 511
mu_B = physical_constants['Bohr magneton'][0]
# Define the parameters
m_star = 0.05 # effective mass in units of m0
g = 4 # g-factor
B = 1 # magnetic field in Tesla
eta = 10e-11 - m # momentum scattering rate in eV-m
nu = 2*eta # spin-flip scattering rate in eV-m
hbar = 1.054e-34
# Define the range of wavevectors
kx = np.linspace(-1e7, 1e7, 501) # 501 points from -1e7 to 1e7 cm^-1
ky = np.linspace(-1e7, 1e7, 501) # 501 points from -1e7 to 1e7 cm^-1
kx, ky = np.meshgrid(kx, ky)

# Calculate the energy levels for spin-up and spin-down states
E_up = (hbar**2 * (kx**2 + ky**2)) / (2 * m_star) + g * hbar * B * np.sqrt(kx**2 + ky**2) - eta * (hbar**2 * (kx**2 + ky**2) / m_star) / (hbar * np.sqrt(kx**2 + ky**2) / m_star + nu)
E_down = (hbar**2 * (kx**2 + ky**2)) / (2 * m_star) - g * hbar * B * np.sqrt(kx**2 + ky**2) - eta * (hbar**2 * (kx**2 + ky**2) / m_star) / (hbar * np.sqrt(kx**2 + ky**2) / m_star + nu)

# Plot the energy dispersion relation
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(kx, ky, E_up, cmap='cool', alpha=0.7)
ax.plot_surface(kx, ky, E_down, cmap='autumn', alpha=0.7)
ax.set_xlabel('kx (cm$^{-1}$)')
ax.set_ylabel('ky (cm$^{-1}$)')
ax.set_zlabel('Energy (eV)')
plt.show()
```

```python
import numpy as np
from scipy.constants import physical_constants
import matplotlib.pyplot as plt

mu_B = physical_constants['Bohr magneton'][0]
m =0.51
# Define the parameters
m_star = 0.05 # effective mass in units of m0
g = 4 # g-factor
B = 1 # magnetic field in Tesla
eta = 10e-11 - m # momentum scattering rate in eV-m
nu = 10e-11 # spin-flip scattering rate in eV-m
hbar = 1.054e-34

# Define the range of wavevectors
kx = np.linspace(-1e17, 1e17, 50) # 50 points from -1e7 to 1e7 cm^-1
ky = np.linspace(-1e17, 1e17, 50) # 50 points from -1e7 to 1e7 cm^-1
kx, ky = np.meshgrid(kx, ky)

# Define the energy function for the in-plane component
def energy_in_plane(k):
    return hbar**2 * k**2 / (2 * m_star)

# Define the energy function for the out-of-plane component
def energy_out_of_plane(k):
    return g * mu_B * B * k

# Define the function to calculate the energy levels
def energy_levels(kx, ky):
    k = np.sqrt(kx**2 + ky**2)
    e_in_plane_up = energy_in_plane(k) + eta * k**2 / (hbar**2 / m_star + nu)
    e_in_plane_down = energy_in_plane(k) - eta * k**2 / (hbar**2 / m_star + nu)
    e_out_of_plane_up = energy_out_of_plane(k)
    e_out_of_plane_down = -energy_out_of_plane(k)
    return e_in_plane_up + e_out_of_plane_up, e_in_plane_down + e_out_of_plane_down

# Calculate the energy levels for spin-up and spin-down states
E_up, E_down = energy_levels(kx, ky)
print(E_up)
print(E_down)

# Plot the energy dispersion relation
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(kx, ky, E_up, cmap='cool', alpha=0.7)
ax.plot_surface(kx, ky, E_down, cmap='autumn', alpha=0.7)
ax.set_xlabel('kx (cm$^{-1}$)')
ax.set_ylabel('ky (cm$^{-1}$)')
ax.set_zlabel('Energy (eV)')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.constants import physical_constants

mu_B = physical_constants['Bohr magneton'][0]

# Define the parameters
m_star = 0.05 # effective mass in units of m0
g = 4 # g-factor
B = 0 # magnetic field in Tesla
eta = 1e-11 # momentum scattering rate in eV-m
nu = 2 * eta # spin-flip scattering rate in eV-m
hbar = 1.054e-34
L = 1e-6 # size of the sample in meters

# Define the range of wavevectors
N = 501
kx = np.linspace(-np.pi/L, np.pi/L, N) 
ky = np.linspace(-np.pi/L, np.pi/L, N) 
kx, ky = np.meshgrid(kx, ky)

# Calculate the energy levels for spin-up and spin-down states
def energy(kx, ky):
    k = np.sqrt(kx**2 + ky**2)
    H0 = hbar**2 * k**2 / (2 * m_star)
    H1 = g * mu_B * B * k / hbar
    Hso = nu * (ky**2 - kx**2) / (2 * m_star)
    return H0 + H1 + Hso

E_up = energy(kx, ky)
E_down = energy(kx, -ky)

# Plot the energy dispersion relation
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(kx, ky, E_up, cmap='cool', alpha=0.7)
ax.plot_surface(kx, ky, E_down, cmap='autumn', alpha=0.7)
ax.set_xlabel('kx (1/m)')
ax.set_ylabel('ky (1/m)')
ax.set_zlabel('Energy (eV)')
plt.show()
```
